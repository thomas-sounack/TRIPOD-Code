```python
import pandas as pd
```

```python
df_TRIPOD = pd.read_csv("TRIPOD/TRIPOD_combined_references.csv")
df_TRIPOD_AI = pd.read_csv("TRIPOD_AI/TRIPOD_AI_combined_references.csv")
```

```python
df_combined = pd.concat([df_TRIPOD, df_TRIPOD_AI], ignore_index=True)
```

```python
def load_PMIDs(path):
    with open(path, "r") as f:
        return {int(line.strip()) for line in f if line.strip()}
    
TRIPOD_PMIDs = load_PMIDs("TRIPOD/TRIPOD_PMIDs.txt")
TRIPOD_AI_PMIDs = load_PMIDs("TRIPOD_AI/TRIPOD_AI_PMIDs.txt")

exclude_PMIDs = TRIPOD_PMIDs.union(TRIPOD_AI_PMIDs)
```

```python
before = df_combined.shape[0]
mask = df_combined["PMID"].isin(exclude_PMIDs)
rows_dropped = df_combined[mask]
```

```python
df_combined = df_combined[~mask].reset_index(drop=True)
after = df_combined.shape[0]
```

```python
print(f"Initial rows: {before}")
print(f"Rows dropped (PMID in PMID lists): {before - after}")
print(f"Final rows: {after}")
```

```python
df_combined.to_csv("merged_references.csv", index=False)
```
