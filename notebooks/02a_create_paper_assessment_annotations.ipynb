# Generate Paper Assessment Annotations

This notebook splits the annotated dataset into a dev set (250 samples) and test set (250 samples)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df_annotations = pd.read_csv("../data/02_paper_assessment/gt_paper_assessment.csv")
```

```python
# Randomly split into two halves
dev, test = train_test_split(df_annotations, test_size=0.5, random_state=42)

# Reset index for both halves
dev = dev.reset_index(drop=True)
test = test.reset_index(drop=True)
```

```python
dev.to_csv("../data/02_paper_assessment/gt_paper_assessment_dev.csv")
test.to_csv("../data/02_paper_assessment/gt_paper_assessment_test.csv")
```
