# Generate Paper Assessment Annotations

This notebook
1. Creates an annotation subset
2. Generates the resources needed to annotate papers for these three questions in Label Studio:
    - is this paper in the scope of our study?
    - if yes, does the paper report a code repository where they are sharing their code?
    - if yes, provide the link to it
3. Based on the individual annotations, generates a second set of annotation resources for reconciliation
4. Organizes the annotation into a gold standard dev set (250 samples) and test set (250 samples)
