```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score

# --- Load data ---
df_to_annotate = df_to_annotate  # already in memory
df_TS = pd.read_csv("../data/7_annotations_repos/TS_repo_annotations.csv")
df_RG = pd.read_csv("../data/7_annotations_repos/RG_repo_annotations.csv")

# --- Columns that correspond to actual annotation questions ---
question_cols = [
    "repo_empty",
    "readme_included",
    "readme_overview",
    "requirements_included",
    "dependency_versions",
    "license_included",
    "coding_conventions",
    "linter_formatter",
    "inline_documentation",
    "modularity",
    "unit_testing",
    "ci_framework",
    "seed_reproducibility",
    "hardware_requirements",
    "reporting_guidance",
    "link_to_paper",
    "citation_file",
    "api_version_documented",
    "sample_dataset",
    "dataset_link",
    "confidential_data_excluded",
    "env_variables_used",
]

# --- Restrict each annotator to repos we care about (optional but usually desired) ---
pmcid_set = set(df_to_annotate["PMCID"])
df_TS_sub = df_TS[df_TS["PMCID"].isin(pmcid_set)].copy()
df_RG_sub = df_RG[df_RG["PMCID"].isin(pmcid_set)].copy()

# --- Inner join on PMCID + repo_url to get only repos both annotated ---
df_both = df_TS_sub.merge(
    df_RG_sub,
    on=["PMCID", "repo_url"],
    suffixes=("_TS", "_RG"),
    how="inner",
)

print(f"Total repos to annotate: {len(df_to_annotate)}")
print(f"TS annotated: {df_TS_sub['PMCID'].nunique()} unique PMCIDs")
print(f"RG annotated: {df_RG_sub['PMCID'].nunique()} unique PMCIDs")
print(f"Overlap (both annotated): {df_both['PMCID'].nunique()} unique PMCIDs\n")

# --- Build long-form side-by-side table for all questions (optional but handy) ---
rows = []
for q in question_cols:
    ts_col = f"{q}_TS"
    rg_col = f"{q}_RG"
    if ts_col not in df_both.columns or rg_col not in df_both.columns:
        continue

    for _, row in df_both[["PMCID", "repo_url", ts_col, rg_col]].iterrows():
        rows.append(
            {
                "PMCID": row["PMCID"],
                "repo_url": row["repo_url"],
                "question": q,
                "TS_label": row[ts_col],
                "RG_label": row[rg_col],
            }
        )

df_side_by_side = pd.DataFrame(rows)

# print("Example of side-by-side annotations (first 10 rows):")
# print(df_side_by_side.head(10), "\n")

# --- Per-question stats + agreement ---
summary_rows = []

for q in question_cols:
    ts_col = f"{q}_TS"
    rg_col = f"{q}_RG"
    if ts_col not in df_both.columns or rg_col not in df_both.columns:
        continue

    # Non-null counts per annotator
    n_TS = df_both[ts_col].notna().sum()
    n_RG = df_both[rg_col].notna().sum()

    # Only rows where both annotators provided a label
    sub = df_both[[ts_col, rg_col]].dropna()
    n_both = len(sub)

    if n_both > 0:
        # Raw agreement
        agreement = (sub[ts_col] == sub[rg_col]).mean()

        # Cohen's kappa (handle degenerate cases)
        try:
            kappa = cohen_kappa_score(sub[ts_col], sub[rg_col])
        except Exception:
            kappa = float("nan")
    else:
        agreement = float("nan")
        kappa = float("nan")

    # Label distributions
    ts_counts = df_both[ts_col].value_counts(dropna=False)
    rg_counts = df_both[rg_col].value_counts(dropna=False)

    print(f"=== Question: {q} ===")
    print("TS label counts:")
    print(ts_counts)
    print("\nRG label counts:")
    print(rg_counts)
    print(f"\nN_TS (non-null): {n_TS}")
    print(f"N_RG (non-null): {n_RG}")
    print(f"N_both (for agreement): {n_both}")
    print(f"Percent agreement: {agreement:.3f}" if n_both > 0 else "Percent agreement: N/A")
    print(f"Cohen's kappa: {kappa:.3f}" if n_both > 0 else "Cohen's kappa: N/A")
    print("-" * 60)

    summary_rows.append(
        {
            "question": q,
            "N_TS": n_TS,
            "N_RG": n_RG,
            "N_both": n_both,
            "agreement": agreement,
            "kappa": kappa,
        }
    )

summary_df = pd.DataFrame(summary_rows).sort_values("question")
print("\n=== Summary table (per-question agreement) ===")
print(summary_df)
```

```python
def compare_annotation_column(colname, df_TS=df_TS, df_RG=df_RG):
    """
    Display rows where TS and RG disagree for a given annotation column.
    colname should be one of the annotation fields, e.g., 'repo_empty'.
    """
    # PMCIDs annotated by both
    pmcids_both = set(df_TS["PMCID"]) & set(df_RG["PMCID"])

    # Filter to overlapping annotations
    TS_both = df_TS[df_TS["PMCID"].isin(pmcids_both)]
    RG_both = df_RG[df_RG["PMCID"].isin(pmcids_both)]

    # Merge side-by-side
    df_compare = TS_both[["PMCID", "repo_url", colname]].merge(
        RG_both[["PMCID", "repo_url", colname]],
        on=["PMCID", "repo_url"],
        suffixes=("_TS", "_RG")
    )

    # Keep only disagreements
    df_disagree = df_compare[
        df_compare[f"{colname}_TS"] != df_compare[f"{colname}_RG"]
    ].sort_values("PMCID")

    return df_disagree
```
