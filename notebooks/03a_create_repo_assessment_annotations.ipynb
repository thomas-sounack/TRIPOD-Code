```python
import pandas as pd
from tqdm.auto import tqdm
import sys
sys.path.append("../")
tqdm.pandas()

from src.repo_utils import clone_and_extract_tree
```

```python
# gpt 5.2's context window is 400k tokens, leave margin for prompt + tree
CONTEXT_WINDOW = 380000
```

```python
df_annotations = pd.read_csv("../data/03_repo_assessment/raw_annotations_repo_assessment.csv")
```

```python
def wrapper_clone_and_extract_tree(repo_url):
    result = clone_and_extract_tree(
        repo_url=repo_url,
        context_window=CONTEXT_WINDOW,
        clone_dir="../data/03_repo_assessment/cloned_repos",
    )
    return {
        "repo_content": result.output,
        "repo_status": result.status,
        "repo_error": result.error,
    }
```

```python
df_annotations[["repo_content", "repo_status", "repo_error"]] = (
    df_annotations["repo_url"]
    .progress_apply(wrapper_clone_and_extract_tree)
    .apply(pd.Series)
)
```

```python
# Rename columns by adding 'gt_' prefix to the specified columns
cols_to_rename = {
    'is_empty': 'gt_is_empty',
    'contains_readme': 'gt_contains_readme',
    'readme_purpose_and_outputs': 'gt_readme_purpose_and_outputs',
    'contains_requirements': 'gt_contains_requirements',
    'requirements_dependency_versions': 'gt_requirements_dependency_versions',
    'contains_license': 'gt_contains_license',
    'sufficient_code_documentation': 'gt_sufficient_code_documentation',
    'is_modular_and_structured': 'gt_is_modular_and_structured',
    'implements_tests': 'gt_implements_tests',
    'fixes_seed_if_stochastic': 'gt_fixes_seed_if_stochastic',
    'lists_hardware_requirements': 'gt_lists_hardware_requirements',
    'contains_link_to_paper': 'gt_contains_link_to_paper',
    'contains_citation': 'gt_contains_citation',
    'includes_data_or_sample': 'gt_includes_data_or_sample'
}

df_annotations.rename(columns=cols_to_rename, inplace=True)
```
